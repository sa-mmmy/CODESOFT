data cleaning 

```python
# Drop irrelevant columns
data_train = data_train.drop(columns=['first', 'last', 'street', 'cc_num', 'trans_num', 'job', 'state', 'city','merchant'])
data_test = data_test.drop(columns=['first', 'last', 'street', 'cc_num', 'trans_num', 'job', 'state', 'city','merchant'])
```

```python
# Convert dates to useful features
for df in [data_train, data_test]:
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
    df['transaction_year'] = df['trans_date_trans_time'].dt.year
    df['transaction_month'] = df['trans_date_trans_time'].dt.month
    df['transaction_day'] = df['trans_date_trans_time'].dt.day
    df['transaction_hour'] = df['trans_date_trans_time'].dt.hour
    df['dob'] = pd.to_datetime(df['dob'])
    df['age'] = df['transaction_year'] - df['dob'].dt.year
    df.drop(columns=['trans_date_trans_time', 'dob'], inplace=True)
```

```python
# One-Hot Encoding for categorical variables
categorical_columns = [ 'category', 'gender']
X_train = pd.get_dummies(data_train.drop(columns=['is_fraud']), columns=categorical_columns)
y_train = data_train['is_fraud']
X_test = pd.get_dummies(data_test.drop(columns=['is_fraud']), columns=categorical_columns)
y_test = data_test['is_fraud']
```

```python
n_c=[X_train.shape[1] , X_test.shape[1]]
n_c
```

```python
# Align columns of test set with training set
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
# Feature Scaling
numerical_columns = ['amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'transaction_year', 'transaction_month', 'transaction_day', 'transaction_hour', 'age']
scaler = StandardScaler()
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.transform(X_test[numerical_columns])
```

```python
X_train
```

```python
# Balance the training data with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Initialize the Logistic Regression model
log_reg = LogisticRegression(solver='saga', class_weight='balanced', random_state=42)
```

```python
# Train the model
log_reg.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred_log_reg = log_reg.predict(X_test)

# Evaluate the model
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log_reg))
print(f"Accuracy: {accuracy_score(y_test, y_pred_log_reg)}")
```
